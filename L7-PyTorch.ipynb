### Step 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define number of weeks in a year
weeks = np.arange(1, 53)

# Starting number of commits per week
initial_commits = 10

# Exponential growth in commits
exp_commits = initial_commits * (1.05 ** (weeks - 1))

# Logistic growth in commits
L = 50
k = 0.2
x0 = 26
logistic_commits = L / (1 + np.exp(-k * (weeks - x0)))

# Linear growth in commits
linear_commits = initial_commits + 0.5 * (weeks - 1)

# Cumulative totals
cumulative_exp = np.sum(exp_commits)
cumulative_logistic = np.sum(logistic_commits)
cumulative_linear = np.sum(linear_commits)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(weeks, exp_commits, label=f'Exponential (total={cumulative_exp:.0f})', linestyle='--')
plt.plot(weeks, logistic_commits, label=f'Logistic (total={cumulative_logistic:.0f})')
plt.plot(weeks, linear_commits, label=f'Linear (total={cumulative_linear:.0f})')
plt.axhline(y=L, color='gray', linestyle=':', label='Upper Limit (50 commits/week)')

plt.title("Weekly Git Commits Over One Year")
plt.xlabel("Week")
plt.ylabel("Commits per Week")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
