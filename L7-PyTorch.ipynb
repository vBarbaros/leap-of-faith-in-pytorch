```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convert_to_excalibur_style_from_array(image_array):
    """
    Converts an image array to a black-and-white sketch style similar to Excalibur drawing tool.

    Parameters:
    - image_array (numpy.ndarray): Input image array.

    Returns:
    - sketch_image: The processed image in Excalibur-like style.
    """
    # Ensure image is grayscale
    if len(image_array.shape) == 3:  # If RGB, convert to grayscale
        image_array = cv2.cvtColor(image_array, cv2.COLOR_RGBA2GRAY)

    # Invert the image
    inverted_image = 255 - image_array

    # Apply Gaussian Blur
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)

    # Create sketch effect
    sketch_image = cv2.divide(image_array, 255 - blurred_image, scale=256)

    # Display the processed image
    plt.figure(figsize=(10, 8))
    plt.imshow(sketch_image, cmap="gray")
    plt.axis("off")  # Keep axes visible
    plt.title("Excalibur-Style Black & White Sketch")
    plt.show()

    return sketch_image

def capture_image_from_plot():
    """
    Captures the currently displayed Matplotlib figure as an image array while keeping the axes.
    """
    # Create a figure and axes
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Example plot (with axes)
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    ax.plot(x, y, color='black', linewidth=2)
    ax.set_xlabel("X-axis Label")  # Keep axis labels
    ax.set_ylabel("Y-axis Label")
    ax.set_title("Original Plot with Axes")

    fig.canvas.draw()

    # Convert figure to an image array (RGBA format)
    img_array = np.array(fig.canvas.renderer.buffer_rgba())

    # Close the figure to free memory
    plt.close(fig)

    return img_array

# Capture the image from the plot (with axes)
image_from_plot = capture_image_from_plot()

# Convert to Excalibur-style sketch while keeping axes
convert_to_excalibur_style_from_array(image_from_plot)
```
